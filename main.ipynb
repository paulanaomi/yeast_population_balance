```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

### Initial population

Creating initial population, considering a normal distribution

```python
num_cells = 100
mu, sigma = 5.1, 0.7 # mean and standard deviation of yeast cell distribution
np.random.seed(42)
cells = np.random.normal(mu, sigma, num_cells)
```

Checking population distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=cells)
plt.title('Distribution of yeast population')
plt.xlabel('Cell size (microns)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Creating a spacial plot for the cells using randomly generated numbers as coordinates

```python
import random
import pandas as pd
import math
```

```python
max_cell_size = math.ceil(max(cells))
max_axis_value = max_cell_size * num_cells + 1
```

```python
random.seed(10)
x = [random.randint(0,max_axis_value) for _ in range(num_cells)]
y = [random.randint(0,max_axis_value) for _ in range(num_cells)]
c = [round(float(cell),1) for cell in cells]
df = pd.DataFrame(zip(x,y,c),columns=['x','y','cell_size'])
df
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(df, x='x', y='y', size='cell_size', hue='cell_size')
plt.title('Distribution of yeast population')
plt.grid(True)
plt.show()
```
