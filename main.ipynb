```python
import matplotlib.pyplot as plt
import seaborn as sns
from Population import *
```

### Initial population

Defining properties of initial population

```python
np.random.seed(42)
random.seed(10)
```

```python
num_cells = 100
```

Create instance of initial population

```python
pop = Population(num_cells)
```

```python
pop.print_cell(0)
```

Return dataframe of inital population

```python
df = pop.return_df()
df.head()
```

Checking population distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df["size"])
plt.title('Distribution of yeast population')
plt.xlabel('Cell size (microns)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

Creating a spacial plot for the cells using randomly generated numbers as coordinates

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(df, x='x', y='y', size='size', hue='size')
plt.title('Distribution of yeast population')
plt.grid(True)
plt.show()
```
